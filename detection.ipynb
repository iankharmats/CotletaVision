```python
import matplotlib.pyplot as plt
import os
```

```python
def get_class_name(idx):
    dir = r"C:\Users\user\Desktop\Python\accelerator2025\data\main_project_data\obj.names" 
    with open(dir, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        return lines[idx].strip()
```

```python
def gist(data):        
    sorted_data = dict(sorted(data.items(), key=lambda item: item[1]))
    _, ax = plt.subplots(figsize=(8, 6))
    ax.barh(list(sorted_data.keys()), list(sorted_data.values()), height=1.5)
    ax.set_axisbelow(True) 
    ax.grid(axis='x', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.title("Распределение классов")
    plt.ylabel("Класс")
    plt.xlabel("Количество")    
    plt.show()
```

```python
vals = {}
dir = r"data\main_project_data\obj_train_data"   
txt_files = [f.split('.')[0] for f in os.listdir(dir) if f.endswith('.txt')]

for txt in txt_files:
    with open(fr'data\main_project_data\obj_train_data\{txt}.txt',
              'r', encoding='utf-8') as file:
        for line in file.readlines():
            cls = line[:line.index(" ")]
            if cls not in vals:
                vals[cls] = 0
            vals[cls] += 1
         
vals_cls = {}   
for idx in vals:
    vals_cls[get_class_name(int(idx))] = vals[idx]
    
gist(vals_cls)
```
